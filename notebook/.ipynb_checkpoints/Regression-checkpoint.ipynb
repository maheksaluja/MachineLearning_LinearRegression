# 1.b) Exploring the dataset

```python
import pandas as pd
import numpy as np
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.model_selection import train_test_split
import statsmodels.api as stm
import statsmodels.formula.api as smf

data = pd.read_excel("../data/CCPP/Folds5x2_pp.xlsx")
rows, column = data.shape
print("No of rows : ", rows)
print("No of columns : ", column)
print("rows represent the number of instance captured")
print('columns represent the feature in a particular observation.')
```

```python
sb.pairplot(data, kind='scatter')
```

```python
for i, col in enumerate(data.columns):
    if(col == "PE"):
        continue
    plt.figure(i)
    sb.boxplot(data = data,y=str(col))
```

### 1.c) Observation : I observe there are outliers in AP and RH which might get removed. And also Observing scatter plot of all predictor against PE, it seems some have linear relation like in scatter plot of AT v/s PE, as  increases, PE decrease. Similar pattern observed in V v/s PE.

# 1.b.iii) Mean, Median, Range, First Quartile, Third Quartile, Interquatile Range

```python
pd.DataFrame({
    "Mean" : data.mean(),
    "Median": data.median(),
    "Range": zip(data.min(), data.max()),
    "First Quartile": data.quantile(.25),
    "Third Quartile": data.quantile(.75),
    "Interquartile Range": data.quantile(.75)-data.quantile(.25)
}).T
```

# 1.c) A simple Linear Regression Model

```python
single_coeff = []
for i,col in enumerate(data.columns):
    if col != 'PE':
        plt.figure(i)
        model = LinearRegression().fit(pd.DataFrame(data[col]), data['PE'])
        prediction = model.predict(pd.DataFrame(data[col]))
        single_coeff.extend(model.coef_)
        plt.scatter(pd.DataFrame(data[col]), data['PE'])
        plt.plot(pd.DataFrame(data[col]), prediction, color='yellow')
        plt.ylabel("Output")
        plt.xlabel(str(col))
```

### 1.c) Observation : AT(Temperature) and V(Exhaust Vaccum) variables are tightly connected with output

# 1.d) Multiple Regression and Analysis

```python
x = data[['AT', 'V', 'AP', 'RH']]
y = data[['PE']]

# model = LinearRegression().fit(x,y)
# pred = model.predict(x)
x = stm.add_constant(x)
stat_model = stm.OLS(y,x).fit()
pred_stat = stat_model.predict(x)
print("-----Coefficients------")
print(stat_model.params)
print("===============================")
print(pd.DataFrame({
    'Actual' : list(y['PE']),
    'Predicted Stats' : list(pred_stat)
}))

stat_model.summary()
```

### 1.d) Analysis :-  All p-values are less than 0.05, which means all predictors impose significant impose on response some how, so we reject the null hypothesis H0.


# 1.e) Coefficient plot from 1.c and 1.d

```python
label = ['AT', 'V', 'AP', 'RH']
for x,y,l in zip(np.array(single_coeff), np.array(stat_model.params[1:]), label):
    plt.scatter(x, y, label= l)
plt.legend()
```

# 1.f) Nonlinear association between any of the predictors and the response

```python
poly = PolynomialFeatures(degree=3)
for i,col in enumerate(data.columns):
    if col != 'PE':
        plt.figure(i)
        x_poly = poly.fit_transform(pd.DataFrame(data[col]))
        model = LinearRegression().fit(x_poly, data['PE'])
        prediction = model.predict(x_poly)
        plt.scatter(pd.DataFrame(data[col]), data['PE'])
        plt.scatter(pd.DataFrame(data[col]), prediction, color='yellow')
        plt.ylabel("Output")
        plt.xlabel("Predicted " + str(col))
```

### 1.f) Analysis :- Based on analysis... AP has high coefficient and more flexible

# 1.g) Full linear regression model with all pairwise interaction

```python
x = data[['AT', 'V', 'AP', 'RH']]
y = data[['PE']]
poly = PolynomialFeatures(2, include_bias=False, interaction_only=True)
x = poly.fit_transform(x)
#print(x[0])
x = stm.add_constant(x)
model = stm.OLS(y, x).fit()
print(model.params)
model.summary()
```

# 1.h) Model using possible interaction terms or nonlinear associations between the predictors and response

```python
x = data[['AT', 'V', 'AP', 'RH']]
y = data[['PE']]
poly = PolynomialFeatures(2, include_bias=False)
x = poly.fit_transform(x)
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3, random_state=70)
#poly = PolynomialFeatures(2, include_bias=False)
#x = poly.fit_transform(train_x)
train_x = stm.add_constant(train_x)
test_x = stm.add_constant(test_x)
model = stm.OLS(train_y, train_x).fit()
y_train_pred = model.predict(train_x)
y_test_pred = model.predict(test_x)
#print(model.params)
model.summary()
#print(pd.DataFrame(train_x))

```

```python
add = 0
for i, j in zip(list(train_y['PE']), list(y_train_pred)):
    add += (i -j)**2
        
print("\n\n =============================\n\n")
print("Train MSE = ", add/len(list(train_y['PE'])))


add = 0
for i, j in zip(list(test_y['PE']), list(y_test_pred)):
    add += (i -j)**2


print("\n\n =============================\n\n")
print("Test MSE = ", add/len(test_y['PE']))
```

### P-Values which are above 0.05 and to be excluded

```python
drop= []
for i in range(4, len(model.pvalues)):
    if model.pvalues[i] > 0.05:
        print(i)
        drop.append(i)
```

```python
train_data_df = pd.DataFrame(train_x)
test_data_df = pd.DataFrame(test_x)
new_train = np.array(train_data_df.drop(train_data_df.columns[drop], axis=1))
new_test = np.array(test_data_df.drop(test_data_df.columns[drop], axis=1))
#print(pd.DataFrame(new_train))
model = stm.OLS(train_y, new_train).fit()
y_train_pred = model.predict(new_train)
y_test_pred = model.predict(new_test)
#print(model.params)
model.summary()
```

```python
add = 0
for i, j in zip(list(train_y['PE']), list(y_train_pred)):
    add += (i -j)**2
        
print("\n\n =============================\n\n")
print("Train MSE = ", add/len(list(train_y['PE'])))


add = 0
for i, j in zip(list(test_y['PE']), list(y_test_pred)):
    add += (i -j)**2


print("\n\n =============================\n\n")
print("Test MSE = ", add/len(test_y['PE']))
```

# 1.i) KNN Regression on Raw Data

```python
from sklearn.neighbors import KNeighborsRegressor
x = data[['AT', 'V', 'AP', 'RH']]
y = data[['PE']]

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3, random_state=70)

train_error = []
test_error = []
k=[]
min_test_error = 1000
min_train_error = 1000
best_k_train = 0
best_k =0

for a in range(1,101, 1):
    
    k.append(1/a)
    knn = KNeighborsRegressor(n_neighbors = a)
    knn.fit(train_x, train_y)
    train_predict = knn.predict(train_x)
    add = 0
    for i, j in zip(list(train_y['PE']), list(train_predict)):
        add += (i -j)**2
    error_train = add/len(list(train_y['PE']))
    train_error.append(add/len(list(train_y['PE'])))
    
    test_predict = knn.predict(test_x)
    add = 0
    for i, j in zip(list(test_y['PE']), list(test_predict)):
        add += (i -j)**2
    error_test = add/len(list(test_y['PE']))
    #print(error_test)
    test_error.append(error_test)
    
    if error_test<min_test_error:
        best_k = a
        min_test_error = error_test
        
    if error_train<min_train_error:
        best_k_train = a
        min_train_error = error_train

plt.plot(k, test_error, label = 'Test Error') 
plt.plot(k, train_error, label = 'Train Error')
plt.xlabel('1/k')
plt.ylabel('Error')
plt.legend()


print('\n\n =======================')
print('Best k for test on raw data  : ', best_k)
print('Best test error on raw data : ', min_test_error)

print('\n\n =======================')
print('Best k for train on raw data  : ', best_k_train)
print('Best train error on raw data : ', min_train_error)
```

# 1.i) KNN Regression with Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data)
normalized_data =pd.DataFrame(scaler.transform(data), columns = ['AT', 'V', 'AP', 'RH', 'PE'])
```

```python
x = normalized_data[['AT', 'V', 'AP', 'RH']]
y = normalized_data[['PE']]

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3, random_state=15)

train_error = []
test_error = []
k=[]
min_test_error = 1000
best_k =0
min_train_error = 1000
best_k_train = 0

for a in range(1,101, 1):
    k.append(1/a)
    knn = KNeighborsRegressor(n_neighbors = a)
    knn.fit(train_x, train_y)
    train_predict = knn.predict(train_x)
    add = 0
    for i, j in zip(list(train_y['PE']), list(train_predict)):
        add += (i -j)**2
    train_error.append(add/len(list(train_y['PE'])))
    
    test_predict = knn.predict(test_x)
    add = 0
    for i, j in zip(list(test_y['PE']), list(test_predict)):
        add += (i -j)**2
    error_test = add/len(list(test_y['PE']))
    #print(error_test)
    test_error.append(error_test)
    
    if error_test<min_test_error:
        best_k = a
        min_test_error = error_test
        
    if error_train<min_train_error:
        best_k_train = a
        min_train_error = error_train

plt.plot(k, test_error, label='Test Error') 
plt.plot(k, train_error, label='Train Error')
plt.xlabel('1/k')
plt.ylabel('Error')
plt.legend()

print('\n\n =======================')
print('Best k on raw data  : ', best_k)
print('Best test error on raw data : ', min_test_error)

print('\n\n =======================')
print('Best k for train on raw data  : ', best_k_train)
print('Best train error on raw data : ', min_train_error)
```
